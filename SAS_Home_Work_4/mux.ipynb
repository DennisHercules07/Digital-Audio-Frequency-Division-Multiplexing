```python
import import_ipynb
import numpy as np
from upconverter import upconverter;

'''
INPUT
    audio: matrix containing audio signals on each column. Note that in the
    exercise pdf, the carrier frequency of index i corresponds to the
    carrier frequency applied to the ith column of the audio matrix.
    Fs: sampling frequency of all audio signals
    F_bw: the multiplexing bandwidth, defined as half the gap
    between any two subsequent carrier frequencies. 
OUTPUT
    mux_signal: the multiplexed signal, with multiplexing bandwidth
    mux_bandwidth
'''

def mux(audio, Fs, F_bw,L):
    # Write your implementation here
    
    y_mux = np.zeros(L)
    for i, y in enumerate(audio):
        
        up = upconverter(i, L, Fs, F_bw)
    
        y_mux += up*audio[i]
    return y_mux.astype('float64')
```
