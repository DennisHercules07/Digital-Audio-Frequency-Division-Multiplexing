```python
import numpy as np

'''
INPUT
    index: column of the desired audio channel
    L: length of the original signal
    Fs: sampling frequency of the audio signal
    F_bw: carrier frequency that you wish to downconvert from.
OUTPUT
    down: the downconverted audio signal
'''

def downconverter(index, L, Fs, F_bw):
    # Write your implementation here
    down = np.zeros(L) 
    fc = F_bw
    
    for n in range(L):
     down[n] = np.exp(-1j*2*np.pi*(fc/Fs)*n)

    return down


```
