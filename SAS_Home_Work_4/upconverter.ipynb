```python
import numpy as np

'''
INPUT
    index: index of the audio file for which this upconverter is for
    L: length of the audio signal
    Fs: sampling frequency of the audio signal
    F_bw: carrier frequency that you wish to upconvert to.
OUTPUT
    up: the upconverted audio signal
'''

def upconverter(index, L, Fs, F_bw):
    # Write your implementation here
    up = np.zeros(L) 
    fc = (2*index+1)*F_bw
    
    for i in range(L):
      up[i] = np.exp(2j*np.pi*(fc/Fs)*i)

    return up


                                  
```
