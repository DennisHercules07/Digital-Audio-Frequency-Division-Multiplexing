```python
import numpy as np
import simpleaudio as sa
import time

'''
INPUT
    audio: Audio signal in the time domain
    Fs: The sample rate in Hz;
    t: desired time to play audio [s]
'''

def play_buffer_wrapper(audio, Fs, t):
    # Normalize to 16-bit range
    audio *= 32767 / np.max(np.abs(audio))
    # Convert to 16-bit data
    audio = audio.astype(np.int16)
    n_channels = 1
    sample_width = 2
    play_obj = sa.play_buffer(audio, n_channels, sample_width, Fs) 
    time.sleep(t)
    play_obj.stop()
    return
```
